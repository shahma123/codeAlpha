```python
import scapy.all as scapy
print(scapy.get_if_list())
```

```python
from scapy.all import sniff, IP, TCP, UDP, Raw

def process_packet(packet):
    if packet.haslayer(IP):
        src_ip = packet[IP].src
        dst_ip = packet[IP].dst
        proto = "TCP" if packet.haslayer(TCP) else "UDP" if packet.haslayer(UDP) else "Other"

        print(f"[+] {src_ip} -> {dst_ip} | Proto: {proto}")

        # Check for data payload
        if packet.haslayer(Raw):
            print(f"    Raw Data: {packet[Raw].load[:30]}")

print("Starting sniffer for 20 packets...")

# We use iface='eth0' (common in Colab) and count=20 so it stops automatically
sniff(iface="eth0", prn=process_packet, count=20, store=0)

print("\nCapture Complete.")
```
